```python
# import library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing Dataset
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
dataset = pd.read_csv('diabetes_prediction_dataset.csv')
dataset['gender'] = lab.fit_transform(dataset['gender'])
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
dataset
```

```python
# Periksa data kosong
X.isnull().any()
```

```python
y.isnull().any()
```

```python
# Membuang data yang kosong
X = X.dropna()
y = y.dropna()
```

```python
# membuat index menjadi berurutan setelah membuang data yang kosong
X.reset_index(drop=True, inplace=True)
y.reset_index(drop=True, inplace=True)
```

```python
# Periksa data duplikat
X.duplicated()
```

```python
X.drop_duplicates()
```

```python
# Periksa data kolom 'age'
column_name = 'age'
non_integer_data = X[column_name].apply(lambda x: not isinstance(x, int))
```

```python
X
```

```python
# Definisikan map dictionary untuk 'smoking_history'
mapping = {
    "No Info": 0,
    "never" : 1,
    "former" : 2,
    "not current" : 3,
    "ever" : 4,
    "current" : 5,
}
X['smoking_history'] = X['smoking_history'].replace(mapping)
dataset.head(5)
```

```python
y = np.array(y).reshape(len(y),1)
```

```python
# Bagi dataset menjadi data latih dan data uji
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test =train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_test
```

```python
print(y_train)
```

```python
print(y_test)
```

Feature Scaling

```python
# regresi model yg membutuhkan fiture scalling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# Training the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred ) #y_test bida dignti dgn y_true
print(cm)
```

```python
accuracy_score(y_test, y_pred)
```

2. Training the KNN model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier2 = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
classifier2.fit(X_train,y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
print(np.concatenate((y_pred2.reshape(len(y_pred2),1), y_test.reshape(len(y_test),1)),1))
```

Making the Confusion Matrix

```python
cm2 = confusion_matrix(y_test,y_pred2)
print(cm2)
```

```python
accuracy_score(y_test,y_pred2)
```

3.Training the SVM model on the Training set

```python
from sklearn.svm import SVC
classifier3 = SVC(kernel='linear', random_state = 0)
classifier3.fit(X_train,y_train)
```

```python
y_pred3 = classifier3.predict(X_test)
print(np.concatenate((y_pred3.reshape(len(y_pred3),1), y_test.reshape(len(y_test),1)),1))
```

```python
cm3 = confusion_matrix(y_test,y_pred2)
print(cm3)
```

```python
accuracy_score(y_test,y_pred3)
```
